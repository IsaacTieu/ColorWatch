```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# If any of the csv files are renamed, make sure to update them here as appropriate
```

```python
df = pd.read_csv('colorData.csv')
print(df)
```

```python
# Plotting change in red

fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Adjust figsize for better layout
x = range(len(df))

axes[0].plot(x, df['Red'], color='red')
axes[0].set_xlabel("Frame Number")
axes[0].set_ylabel("Red Value")
axes[0].set_title("Red Value vs. Frame Number")

axes[1].plot(x, df['Green'], color='green')
axes[1].set_xlabel("Frame Number")
axes[1].set_ylabel("Green Value")
axes[1].set_title("Green Value vs. Frame Number")

axes[2].plot(x, df['Blue'], color='blue')
axes[2].set_xlabel("Frame Number")
axes[2].set_ylabel("Blue Value")
axes[2].set_title("Blue Value vs. Frame Number")

plt.tight_layout()
plt.show()
```

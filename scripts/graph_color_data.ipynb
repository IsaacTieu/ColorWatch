This notebook outlines some simple color graph plotting with Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

# If any of the csv files are renamed, make sure to update them as appropriate
```

```python
# Change based on where this was saved.
colorData_path = 'colorData.csv'

df = pd.read_csv(colorData_path)
print(df)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # Adjust figsize for better layout
x = range(len(df))

axes[0].plot(x, df['Red'], color='red')
axes[0].set_xlabel("Frame Number")
axes[0].set_ylabel("Red Value")
axes[0].set_title("Red Value vs. Frame Number")

axes[1].plot(x, df['Green'], color='green')
axes[1].set_xlabel("Frame Number")
axes[1].set_ylabel("Green Value")
axes[1].set_title("Green Value vs. Frame Number")

axes[2].plot(x, df['Blue'], color='blue')
axes[2].set_xlabel("Frame Number")
axes[2].set_ylabel("Blue Value")
axes[2].set_title("Blue Value vs. Frame Number")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots() 
x = range(len(df))

axes.plot(x, df['Red'], color='red')
axes.plot(x, df['Green'], color='green')
axes.plot(x, df['Blue'], color='blue')
axes.set_xlabel("Frame Number")
axes.set_ylabel("RGB Value")
axes.set_title("RGB Value vs. Frame Number")

plt.tight_layout()
plt.show()
```
